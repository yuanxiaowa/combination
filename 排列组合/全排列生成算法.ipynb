## SJT算法

```python
import numpy as np
```

```python
di = -1
n = 4
arr = np.arange(n)+1
print(arr)
while True:
    canOp = False
    if di is -1:
        for t in range(n - 1):
            i = n-1-t
            arr[[i, i-1]] = arr[[i-1, i]]
            print(arr)
        for t in range(n - 1):
            i = n - 1 - t
            if arr[i] > arr[i - 1]:
                arr[[i, i - 1]] = arr[[i - 1, i]]
                canOp = True
                break
    else:
        for i in range(n - 1):
            arr[[i, i+1]] = arr[[i+1, i]]
            print(arr)
        for i in range(n - 1):
            if arr[i + 1] > arr[i]:
                arr[[i, i + 1]] = arr[[i + 1, i]]
                canOp = True
                break
    if np.all(arr[:-1]>arr[1:]) and not canOp:
        break
    print('---------------')
    print(arr)
    di = -di
```
